```python
import pandas as pd
import numpy as np
profit = []
highs = []
lows = []

df = pd.read_pickle("GBP_USD_CORRECT.pkl")
for eachRecord in df.to_dict(orient='records'):
    #The actual value is less than the forecasted value
    #Price should fall...
    openingPrice = eachRecord["candle_data"]["candles"][0]["mid"]["o"]
    closingPrice = eachRecord["candle_data"]["candles"][0]["mid"]["c"]
    highestPrice = eachRecord["candle_data"]["candles"][0]["mid"]["h"]
    lowestPrice = eachRecord["candle_data"]["candles"][0]["mid"]["h"]
    if float(eachRecord["actual"][:-1]) < float(eachRecord["forecast"][:-1]):
        #The opening price is larger than the closing price, price has fallen
        if openingPrice > closingPrice:
            # print(float(openingPrice) - float(closingPrice))
            # print(pow(10, -4))
            # print("Final", (float(openingPrice) - float(closingPrice)) / pow(10, -4))
            profit.append(round(((float(openingPrice) - float(closingPrice)) / pow(10, -4)), 1))
            lows.append(round(((float(openingPrice) - float(lowestPrice)) / pow(10, -4)), 1))
    
    #The actual value is larger than the forecasted value
    #Price should rise...
    elif float(eachRecord["actual"][:-1]) > float(eachRecord["forecast"][:-1]):
        #The opening price is less than the closing price, price has risen
        if openingPrice < closingPrice:
            # print(float(openingPrice) - float(closingPrice))
            # print(pow(10, -4))
            # print((float(openingPrice) - float(closingPrice)) / pow(10, -4))
            profit.append(round(((float(closingPrice) - float(openingPrice)) / pow(10, -4)), 1))
            highs.append(round(((float(highestPrice) - float(openingPrice)) / pow(10, -4)), 1))

print(len(profit))
print("Maximum:", np.max(profit))
print("Mean:", np.mean(profit))
print("Minimum:", np.min(profit))
print("Total won:", np.sum(profit))
print("Highest high:", np.max(highs))
print("Average high:", np.mean(highs))
print("Lowest low:", np.min(lows))
print("Average low:", np.mean(lows))
print("Amount of trades we correctly expected price to fall:", len(lows))
print("Amount of trades we correctly expected price to rise:", len(highs))
print(highs, end="\n\n")
print(lows)
```

```python
import pandas as pd
import numpy as np

losses = []
#The highest price it went in that hour
highs = []
lows = []

df = pd.read_pickle("GBP_USD_INCORRECT.pkl")

for eachRecord in df.to_dict(orient='records'):
    #The actual value is less than the forecasted value
    #Price should fall...
    openingPrice = eachRecord["candle_data"]["candles"][0]["mid"]["o"]
    closingPrice = eachRecord["candle_data"]["candles"][0]["mid"]["c"]
    highestPrice = eachRecord["candle_data"]["candles"][0]["mid"]["h"]
    lowestPrice = eachRecord["candle_data"]["candles"][0]["mid"]["h"]
    if float(eachRecord["actual"][:-1]) < float(eachRecord["forecast"][:-1]):
        #The opening price is larger than the closing price, price has fallen
        #But remember, because we know that these trade failed, the price movement should be the opposite to our expectations.
        #So instead of the price falling because the evaluation underperformed, the price should rise:
        if openingPrice < closingPrice:
            # print(float(openingPrice) - float(closingPrice))
            # print(pow(10, -4))
            # print("Final", (float(openingPrice) - float(closingPrice)) / pow(10, -4))
            #print(float(eachRecord["actual"][:-1]), float(eachRecord["forecast"][:-1]), openingPrice, closingPrice)
            losses.append(round(((float(openingPrice) - float(closingPrice)) / pow(10, -4)), 1))
            lows.append(round(((float(openingPrice) - float(lowestPrice)) / pow(10, -4)), 1))

    
    #The actual value is larger than the forecasted value
    #Price should rise...
    elif float(eachRecord["actual"][:-1]) > float(eachRecord["forecast"][:-1]):
        #The opening price is less than the closing price, price has risen
        if openingPrice > closingPrice:
            # print(float(openingPrice) - float(closingPrice))
            # print(pow(10, -4))
            # print((float(openingPrice) - float(closingPrice)) / pow(10, -4))
            #print(float(eachRecord["actual"][:-1]), float(eachRecord["forecast"][:-1]), openingPrice, closingPrice)
            losses.append(round(((float(closingPrice) - float(openingPrice)) / pow(10, -4)), 1))
            highs.append(round(((float(highestPrice) - float(openingPrice)) / pow(10, -4)), 1))

    #Yeah this actually happened lmao
    if openingPrice == closingPrice:
        losses.append(0)

print(len(losses))
print("Maximum:", np.max(losses))
print("Mean:", np.mean(losses))
print("Minimum:", np.min(losses))
print("Total loss:", np.sum(losses))
print("Highest high:", np.max(highs))
print("Average high:", np.mean(highs))
print("Lowest low:", np.min(lows))
print("Average low:", np.mean(lows))
print("Amount of trades we correctly expected price to fall:", len(lows))
print("Amount of trades we correctly expected price to rise:", len(highs))
```
